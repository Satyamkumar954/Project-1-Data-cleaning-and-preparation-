```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries imported successfully!")
```

```python
FILE_PATH = r"C:\Users\Satayam Sahu\OneDrive\Desktop\decodelab project 1\DecodeLabs_P1_Dataset.xlsx"

df = pd.read_excel(FILE_PATH, sheet_name="RAW_DATA (Messy)")

print(f"📊 Shape: {df.shape}  (rows x columns)")
print(f"📋 Columns: {list(df.columns)}")
df
```

```python
print("=" * 50)
print("📌 DATA TYPES:")
print(df.dtypes)
print()
print("=" * 50)
print("📌 MISSING VALUES (Null counts):")
print(df.isnull().sum())
print()
print("=" * 50)
print("📌 BASIC STATISTICS:")
df.describe(include='all')
```

```python
# Check for duplicate rows
print(f"🔁 Total duplicate rows: {df.duplicated().sum()}")
print()
print("Duplicate rows detail:")
df[df.duplicated(keep=False)]
```

```python
before = df.shape[0]
df = df.drop_duplicates(keep='first').reset_index(drop=True)
after = df.shape[0]

print(f"✅ ISS001 & ISS002 Fixed!")
print(f"   Before: {before} rows  →  After: {after} rows")
print(f"   {before - after} duplicate row(s) removed")
```

```python
print("City values BEFORE trimming:")
print(df['City'].tolist())

df['City'] = df['City'].str.strip()

print("\nCity values AFTER trimming:")
print(df['City'].tolist())
print("✅ ISS015 Fixed! Leading/trailing whitespace removed")
```

```python
print("Unique city values BEFORE standardization:")
print(df['City'].unique())

# Mapping of incorrect → correct city names
city_mapping = {
    'mumbai' : 'Mumbai',
    'CHENNAI': 'Chennai',
    'BLR'    : 'Bengaluru',
    'Blor'   : 'Bengaluru',
}

df['City'] = df['City'].replace(city_mapping)

print("\nUnique city values AFTER standardization:")
print(df['City'].unique())
print("✅ ISS010 Fixed! City names standardized")
```

```python
print("Names BEFORE:")
print(df['Customer_Name'].tolist())

df['Customer_Name'] = df['Customer_Name'].str.title()

print("\nNames AFTER:")
print(df['Customer_Name'].tolist())
print("✅ ISS009 Fixed! Names converted to Proper Case")
```

```python
print("Category unique values BEFORE:", df['Category'].unique())
print("Status unique values BEFORE  :", df['Status'].unique())

df['Category'] = df['Category'].str.title()
df['Status']   = df['Status'].str.title()  # NaN values are preserved automatically

print("\nCategory unique values AFTER :", df['Category'].unique())
print("Status unique values AFTER   :", df['Status'].unique())
print("✅ ISS011 & ISS012 Fixed! Category and Status standardized")
```

```python
print("Join_Date BEFORE:")
print(df['Join_Date'].tolist())

df['Join_Date'] = pd.to_datetime(df['Join_Date'], dayfirst=True, errors='coerce')
df['Join_Date'] = df['Join_Date'].dt.strftime('%Y-%m-%d')

print("\nJoin_Date AFTER:")
print(df['Join_Date'].tolist())
print("✅ ISS008 Fixed! All dates converted to YYYY-MM-DD format")
```

```python
print("Rows with Age = 999:")
print(df[df['Age'] == 999][['Customer_ID', 'Customer_Name', 'Age']])

# Replace 999 with NaN — will be imputed with median in the next step
df['Age'] = df['Age'].replace(999, np.nan)

print("\n✅ ISS013 Fixed! Age=999 replaced with NaN (will be imputed below)")
```

```python
age_median = df['Age'].median()
missing_age_rows = df[df['Age'].isna()][['Customer_ID', 'Customer_Name']]

print(f"Median Age: {age_median}")
print(f"Rows with missing Age:\n{missing_age_rows}")

df['Age'] = df['Age'].fillna(age_median).astype(int)

print(f"\n✅ ISS004 Fixed! {len(missing_age_rows)} missing Age value(s) filled with median ({age_median})")
```

```python
# Calculate mean using only valid (positive) values
valid_amounts = df[df['Purchase_Amount'] > 0]['Purchase_Amount']
amount_mean = round(valid_amounts.mean(), 2)

missing_amount_rows = df[df['Purchase_Amount'].isna()][['Customer_ID', 'Customer_Name']]
print(f"Mean Purchase_Amount (from valid values): {amount_mean}")
print(f"Rows with missing Purchase_Amount:\n{missing_amount_rows}")

df['Purchase_Amount'] = df['Purchase_Amount'].fillna(amount_mean)

print(f"\n✅ ISS005 Fixed! {len(missing_amount_rows)} missing value(s) filled with mean ({amount_mean})")
```

```python
print("Rows with negative Purchase_Amount:")
print(df[df['Purchase_Amount'] < 0][['Customer_ID', 'Customer_Name', 'Purchase_Amount']])

# Replace negative values with NaN, then fill with mean
df.loc[df['Purchase_Amount'] < 0, 'Purchase_Amount'] = np.nan
mean_val = round(df['Purchase_Amount'].mean(), 2)
df['Purchase_Amount'] = df['Purchase_Amount'].fillna(mean_val)

print(f"\n✅ ISS014 Fixed! Negative value replaced with mean ({mean_val})")
```

```python
print("Rows with missing Customer_Name:")
print(df[df['Customer_Name'].isna()][['Customer_ID', 'Customer_Name']])

df['Customer_Name'] = df['Customer_Name'].fillna('Unknown')

print("✅ ISS003 Fixed! Missing name filled with 'Unknown'")
```

```python
print("Rows with missing Email:")
print(df[df['Email'].isna()][['Customer_ID', 'Customer_Name', 'Email']])

df['Email'] = df['Email'].fillna('Not Provided')

print("✅ ISS006 Fixed! Missing email filled with 'Not Provided'")
```

```python
status_mode = df['Status'].mode()[0]
print(f"Mode of Status column: {status_mode}")
print("Rows with missing Status:")
print(df[df['Status'].isna()][['Customer_ID', 'Customer_Name', 'Status']])

df['Status'] = df['Status'].fillna(status_mode)

print(f"\n✅ ISS007 Fixed! Missing status filled with mode ('{status_mode}')")
```

```python
print("=" * 55)
print("📊 FINAL CLEANED DATASET")
print("=" * 55)
print(f"Shape: {df.shape}")
print()
print("Missing Values Check (should all be 0):")
print(df.isnull().sum())
print()
print(f"Duplicate Rows Check (should be 0): {df.duplicated().sum()}")
print(f"Negative Purchase_Amount Check (should be 0): {(df['Purchase_Amount'] < 0).sum()}")
```
